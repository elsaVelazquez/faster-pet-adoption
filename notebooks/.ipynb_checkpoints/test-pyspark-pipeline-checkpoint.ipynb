```python
import pyspark as ps    # for the pyspark suite
from pyspark.sql.types import *
from datetime import datetime
import json

spark = ps.sql.SparkSession.builder \
            .master("local[4]") \
            .appName("sparkSQL exercise") \
            .getOrCreate()

sc = spark.sparkContext 
```

```python
def apply_json(input_lst):
    try:
        return json.loads(input_lst)
    except:
        return None 
```

```python
data_filepath = '../data/json/api_dump_2020-06.json'

data = sc.textFile(data_filepath).map(apply_json) #reads in the json file
```

```python
data.take(10)
```

```python
def new_rdd(row):
    try:
        
        animals = row['animals']
        id = row['animals'][0]
#         age = row['age']
#         photos = row['photos']
#         status = row['status']
#         status_changed_at = row['status_changed_at']
#         description = row['description']
        
        

        return animals #id, age, photos, status, status_changed_at, description
    
    except:
        return None

rdd_small = data.map(new_rdd)

rdd_small.take(5)
```
